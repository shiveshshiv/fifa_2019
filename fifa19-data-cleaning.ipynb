**Importing Dataset**

```python
data = pd.read_csv('kl.csv', encoding='ISO-8859-1')
data.head()
```

Checking if any row has all values missing

```python
data.isnull().all(axis=1).sum()
```

Rows with more than 50 missing values

```python
data[data.isnull().sum(axis=1) > 50]
```

```python
print(f'Before Deletion {data.shape[0]}')
data = data[data.isnull().sum(axis=1) < 50]

print(f'After Deletion {data.shape[0]}')
```

Check for missing values in columns

```python
data.isnull().sum()
```

Number of missing values per column

```python
data.isnull().sum()
```

% of missing values per column

```python
x = data.isnull().sum()
y = (data.isnull().sum()/data.shape[0]) * 100

z = {'No. of missing values' :x, '% of missing values' :y}

df = pd.DataFrame(z, columns = ['No. of missing values', '% of missing values'])
df.sort_values(by='% of missing values', ascending=False)
```

Deleting the column with 93% missing values

```python
data = data.drop(['Loaned From'], axis=1)
```

Checking datatypes

```python
data.dtypes[data.isnull().any()]
```

Imputing jersey numbers, since they can't be approximated

```python
data['Jersey Number'].fillna('NA', inplace=True)
```

Filling in the missing values

```python
vals_list1 = ['Club', 'Position', 'Joined', 'Contract Valid Until', 'Release Clause']

for each in vals_list1:
    data[each] = data[each].fillna(data[each].mode()[0])
    
    
all_vals = 'LS ST RS LW LF CF RF RW LAM CAM RAM LM LCM CM RCM RM LWB LDM CDM RDM RWB LB LCB CB RCB RB'
vals_list2 = all_vals.split(' ')

for each in vals_list2:
    data[each].fillna(0, inplace=True)
```

Checking if any missing values left

```python
print(data.isnull().sum())
print('\n Total Missing - ', data.isnull().sum().sum())
```

Finding exact location of missing values

```python
bool_series1 = pd.isnull(data['Age'])
data[bool_series1]
```

```python
bool_series2 = pd.isnull(data['Overall'])
data[bool_series1]
```

Removing that row
